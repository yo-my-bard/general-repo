```python
import pandas as pd
import numpy as np
```

```python
events = pd.read_csv("athlete_events.csv")
```

```python
events.head()
```

```python
#Avg age of Gold, Silver, Bronze, No Medals
#events.fillna("None", inplace=True)
events.groupby(by = "Medal")["Age"].mean()
```

```python
#Which sports have the youngest winners on average? Which have the oldest?
#Ambiguous question here. Is winning just a gold medal or any medal? Let's assume winning is any medal.

events_medal = events.replace({"Medal": {"Bronze": "Medal", "Silver": "Medal", "Gold": "Medal"}}) #Grouping all medal values to just medal
events_medal.groupby(["Sport", "Medal"])['Age'].mean().unstack().sort_values("Medal") #Look at top 5, bottom 5
```

```python
#4. Who has won the most medals?
#Assumes names are spelled consistently

events_medal.groupby(["Name", "Medal"])['Medal'].count().unstack().sort_values("Medal", ascending=False)
```

```python
#5. What percentage of athletes have won medals?
#Filtering source = https://stackoverflow.com/a/38258528
len(events_medal.groupby(["Name", "Medal"])['Medal'].count().unstack()["Medal"][lambda x: x.notnull()])/ len(events_medal.groupby(["Name", "Medal"])['Medal'].count().unstack()["Medal"])
```

```python
#6. Which athlete has competed in the most number of events? The least?
events_medal.groupby(["Name", "Event"])['ID'].count().groupby("Name").sum().sort_values() #Big tie for least number of events
```

```python
#7. Which sports have athletes with the highest weight? The lowest?
#Answering using means, limited to athletes/observations who reported weight. appears to be in (KG)
events_medal.groupby("Sport")["Weight"].mean().sort_values()
```

```python
#8. What's the average age of Gymnastics winners before 1970? After 1970?
#Source: https://stackoverflow.com/a/47291304
events_1970 = events_medal.replace({"Year": {x:"After" if x > 1970 else "Before" for x in range(2017)}})
events_1970.groupby(["Year", "Medal", "Sport"])["Age"].mean().unstack()["Gymnastics"].unstack()
```
